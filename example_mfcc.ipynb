```python
import numpy as np
from matplotlib import pyplot as pl

import scipy.io.wavfile as wav
from scikits.talkbox.features import mfcc

%matplotlib inline
```

```python
sample_rate_1, X_1 = wav.read('vowels/nafdlaa.wav')
sample_rate_2, X_2 = wav.read('vowels/namshaa.wav')
```

```python
pl.figure(figsize=(15,6))
pl.subplot(2,1,1)
pl.plot(X_1)
pl.xlim(0, len(X_1))
pl.title('X_1')
pl.grid()
pl.subplot(2,1,2)
pl.plot(X_2)
pl.xlim(0, len(X_2))
pl.title('X_2')
pl.grid()
pl.tight_layout()
```

```python
ceps_1, _, _ = mfcc(X_1, fs=sample_rate_1)
ceps_2, _, _ = mfcc(X_2, fs=sample_rate_2)
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
pl.boxplot(ceps_1)
pl.title('X_1')
pl.grid()
pl.subplot(1,2,2)
pl.boxplot(ceps_2)
pl.title('X_2')
pl.grid()
```
